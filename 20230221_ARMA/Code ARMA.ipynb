```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import arma_generate_sample

ar = np.array([1, 0.5])
ma = np.array([1, -0.5])
np.random.seed(123457)
x =  arma_generate_sample(ar = ar, ma = ma, nsample = 500)
x = pd.Series(x)
x.plot(); 
plt.title('Simulation of an ARMA(1,1) with (-0.5, -0.5)');
plt.show()

ar = np.array([1, 0.5])
ma = np.array([1, 0.5])
np.random.seed(123457)
x =  arma_generate_sample(ar = ar, ma = ma, nsample = 500)
x = pd.Series(x)
x.plot(); 
plt.title('Simulation of an ARMA(1,1) with (-0.5, 0.5)');
plt.show()

ar = np.array([1, -0.9])
ma = np.array([1, -0.5])
np.random.seed(123457)
x =  arma_generate_sample(ar = ar, ma = ma, nsample = 500)
x = pd.Series(x)
x.plot(); 
plt.title('Simulation of an ARMA(1,1) with (0.9, -0.5)');
plt.show()

ar = np.array([1, -0.9])
ma = np.array([1, 0.5])
np.random.seed(123457)
x =  arma_generate_sample(ar = ar, ma = ma, nsample = 500)
x = pd.Series(x)
x.plot(); 
plt.title('Simulation of an ARMA(1,1) with (0.9, 0.5)');
plt.show()
```
